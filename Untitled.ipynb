```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

# Task 1

```python
X_data_2d = pd.read_csv('data/data_2d.csv')
X_mnist = pd.read_csv('data/mnist.csv')
```

```python
def calculate_sse(X, sse, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

```python
sse_data_2d = []
sse_mnist = []

for k in range(1, 11):
    calculate_sse(X_data_2d, sse_data_2d, k)
    calculate_sse(X_mnist, sse_mnist, k)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(range(1, 11), sse_data_2d, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared errors (SSE)')
plt.title('Elbow Method for Optimal K (2D Data)')

plt.subplot(1, 2, 2)
plt.plot(range(1, 11), sse_mnist, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared errors (SSE)')
plt.title('Elbow Method for Optimal K (MNIST Data)')

plt.tight_layout()
plt.show()
```

# Task 2

```python
pca = PCA(n_components=2)
mnist_data_pca = pca.fit_transform(X_mnist)
```

```python
optimal_k_mnist = 2
kmeans_mnist = KMeans(n_clusters=optimal_k_mnist, random_state=42)
mnist_clusters = kmeans_mnist.fit_predict(mnist_data_pca)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=mnist_data_pca[:, 0], y=mnist_data_pca[:, 1], hue=mnist_clusters, palette='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering on MNIST Data (PCA-reduced to 2D)')
plt.legend(title='Cluster')
plt.show()
```

```python
data_2d_pca = pca.fit_transform(X_data_2d)
```

```python
optimal_k_data_2d = 2
kmeans_data_2d = KMeans(n_clusters=optimal_k_data_2d, random_state=42)
data_2d_clusters = kmeans_data_2d.fit_predict(data_2d_pca)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data_2d_pca[:, 0], y=data_2d_pca[:, 1], hue=data_2d_clusters, palette='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering on 2D Data')
plt.legend(title='Cluster')
plt.show()
```
